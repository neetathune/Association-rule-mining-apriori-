```python
import pandas as pd
data=pd.DataFrame({'Milk':[1, 1, 0 ,1 ,1],
                   'Bread':[1,0,1,1,0],
                   'cheese': [1,1,1,0,0],
                   'Apples':[0,1,0,1,0],
                   'Bananas':[0,1,1,0,1]})
data
```

```python
# generate frequent item set using mlxtend and join association rule
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# set the minimum support 0.4
min_support_threshold=0.4
frequent_itemsets=apriori(data,min_support=min_support_threshold,use_colnames=True)
```

```python
frequent_itemsets
```

```python
# generate association rule
```

```python
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=1)
rules
```
